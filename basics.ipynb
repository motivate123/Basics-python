5.  What is mutability python?

  - Immutable types in Python:

- float

- bool

6.  Why are lists mutable, but tuples are immutable


- Lists are mutable:

  - You can change their items (add, remove, or update).

 - Python made lists this way on purpose so they are flexible.

 - Useful when you need to change data often.

- Tuples are immutable:

   - You cannot change their items after creation.

  - Python made tuples this way to make them safe and fast.

  - Useful when you want fixed data (like days of the week).

13.  What is the difference between for and while loops ?

| Feature   | `for` loop                | `while` loop                    |
| --------- | ------------------------- | ------------------------------- |
| Use for   | **Known number of times** | **Repeat until a condition**    |
| Structure | Loops over items or range | Loops while condition is `True` |


   for i in range(3):

    print(i)       # Runs 3 times

while x < 5:

    print(x)       # Runs as long as x < 5

14.  Describe a scenario where a while loop is more suitable than a for loop??

 - Use a while loop when you don't know in advance how many times the loop will run, but you want it to keep running as long as a condition is true.

| Use `for` when...        | Use `while` when...                       |
| ------------------------ | ----------------------------------------- |
| You know how many times  | You **don’t know** how many times         |
| Looping over items/index | Waiting for a condition to become `False` |
